# FermiSee

![image.gif](https://github.com/TRIQS/FermiSee/raw/unstable/doc/tutorial_gen.gif)

To visualize and explore the spectral properties of materials more interactively we developed the WebApp [FermiSee](https://github.com/TRIQS/FermiSee). In this part of the tutorial we will learn how to use FermiSee and explore its functionality. The App can be accessed via a hosted version at the [Flatiron Institute](http://fermisee.flatironinstitute.org). Or locally under: 


FermiSee is started via `python app.py` in the main directory of the App. This is starting a webserver that you can access under [127.0.0.1:8889](127.0.0.1:8889). If you are using FermiSee on your local machine you should also copy the output files `lco_hr.dat`, `lco.wout`, and `simga_fermisee.h5` to your local machine via scp. If you have trouble you can also use the public version of FermiSee here: [fermisee.flatironinstitute.org/](http://fermisee.flatironinstitute.org/) .

1. Load the `lco_hr.dat` and the `lco.wout` from the `data/mlwf` directory in the `TB Hamiltonian` section and calculate the bands
2. calculate the correct `mu` by first entering a guess in the field then pressing the button `calc mu`. As long as the tab displays "updating ...", FermiSee is still working
3. recalculate bands. They should be shifted now
4. change the kpoint path to the same path used in the previous notebook
5. download the config (for safety)
6. expand now the self-energy section by clicking on it
7. upload the `sigma_fermisee.h5` file and click `Calculate A(k,w)`, after a few seconds the spectral function should appear
8. play a bit around with the features to get yourself familiar with the functionality

If anything goes wrong, reload and upload on the top your previously downloaded config file.
